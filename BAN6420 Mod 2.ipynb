```python
import pandas as pd
import os
import zipfile

# Define the file path
file_path = "C:/Users/Dell/Code/Data/Total.csv"

try:
    # Check if file exists
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file {file_path} does not exist")
    
    # Load the salary data into a DataFrame
    salary_data = pd.read_csv(file_path)
   
    # Preview the data
    #print(salary_data.head(n=2))

except PermissionError:
    print(f"Permission denied. Please ensure that:\n"
          f"1. The file is not open in another program\n"
          f"2. You have read permissions for the file\n"
          f"3. Try running your IDE as administrator")
    
except FileNotFoundError as e:
    print(f"Error: {e}")
    
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python



# Convert to a list of dictionaries (one dictionary per employee)
salary_dict = salary_data.to_dict(orient='records')

#Define the function to search for employees by name
def search_employees_by_name(employee_name):
    try:
        # Ensure the data dictionary exists
        if not isinstance(salary_dict, list):  # Verify that data_dict is a list of dictionaries
            raise ValueError("Error: The data dictionary is not properly initialized.")
         # Remove spaces from the input name and convert to lowercase
        search_name = employee_name.replace(" ", "").lower()
        # List to store matching employees
        matching_employees = []

        # Iterate through the list of dictionaries (processed data)
        for employee in salary_dict:
             if 'EmployeeName' in employee:
                # Remove spaces from the employee's name and compare in a case-insensitive manner
                employee_name_processed = employee['EmployeeName'].replace(" ", "").lower()
                if employee_name_processed == search_name:
                    matching_employees.append(employee)  # Add matching employee to the list

        # Check if any matches were found
        if matching_employees:
            return matching_employees  # Return the list of matching employees
        else:
            return f"No employees found with the name '{employee_name}'."
    
    except ValueError as ve:
        return str(ve)  # Handle the case where data_dict is not properly initialized
    except KeyError as ke:
        return f"Error: Missing expected key in the data - {ke}"
    except Exception as e:
        return f"An unexpected error occurred: {e}"  # Handle any other unexpected exceptions



name = input("Type in the Employee Name")  # Type in the name of the employee to search for

# Calling the function to search for an employee by name
result = search_employees_by_name(name)
if isinstance(result, list):  # If the result is a list of matches
    #for idx, employee in enumerate(result, 1):  # Display each match with numbering
        #print(f"Entry {idx}: {employee}")
        # Converting back to data frame
    employee_profile = search_employees_by_name(name)
    employee_profile = pd.DataFrame(employee_profile)

    # Preview the data
    print("\n Preview \n", employee_profile.head(n=2))

    # Save to CSV
    csv_filename = name.replace(" ", "_") + '.csv'
    employee_profile.to_csv(csv_filename, index=False)
        
    # Create a zip folder
    with zipfile.ZipFile('Employee_Profile.zip', 'w') as zipf:
        zipf.write(csv_filename) # Add the CSV file to the ZIP folder
    os.remove(csv_filename)  # Remove the CSV file after adding it to the ZIP folder
    print("\n Employee details successfully exported and zipped.")

else:
    print(result)  # If no matches were found or an error occurred







```
